# **PERCEPTRON FROM SCRATCH**

```python
x=[0.75,0.65,0.80]
w=[0.4,0.3,0.5]
b = -1000
n = 0.01
y = 1
```

```python
z = w[0]*x[0]+w[1]*x[1]+w[2]*x[2] + b
```

```python
if z < 0:
  y_predicted = 0
else:
  y_predicted = 1

print("Predicted output:", y_predicted)
```

```python
loss = y - y_predicted
print("loss:", loss)
```

```python
if loss == 0:
  print("No update required")
else:
  for i in range(len(w)):
    w[i] = w[i] + n*loss*x[i]
```

```python
b = b + n*loss

print("Updated weights:", w)
print("Updated bias:", b)
```
